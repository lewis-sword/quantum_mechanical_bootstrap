# CNN for HiRISE landmark data #
A multiple convolution layer newtork that aims to classify Mars Landmark data using the HiRISE data set. The set is available at https://zenodo.org/records/2538136, with the paths in the code set up accordingly.

```python
######################################## DATA AND PACKAGE IMPORT ##########################################
import datetime
import numpy as np
import pandas as pd
import random
import os
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix

# Sets seeds. Leads to a deterministic output from running the same model again
seed_num = 17
os.environ['TF_DETERMINISTIC_OPS'] = '1'
os.environ['PYTHONHASHSEED'] = str(seed_num)  # Python general
random.seed(seed_num)  # Python random
np.random.seed(seed_num)
tf.random.set_seed(seed_num)

#Used 7Zip to extract the zip images -- see https://zenodo.org/records/2538136 -- deafult extractor can cause issues
#base_path is the main folder, base_path2 is the image subfolder

base_path = "C:/Users/Lewis Sword/gitTest/hirise_cnn/hirise-map-proj-v3/"
base_path2 = "C:/Users/Lewis Sword/gitTest/hirise_cnn/hirise-map-proj-v3/map-proj-v3"
image_classes = np.loadtxt(base_path + 'labels-map-proj-v3.txt', dtype=str, delimiter=" ")

df_import = pd.DataFrame(image_classes, columns=['filename', 'class'])
# dropping some rows/data from the large "0" class
class_zero_choice = 5000
df_new_1 = df_import.loc[df_import["class"] == "0"][0:class_zero_choice]
df_new_2 = df_import.loc[df_import["class"] != "0"]
df = pd.concat([df_new_1, df_new_2], ignore_index=True)
print(df.groupby(["class"]).count())

# Shuffling the data
df = shuffle(df, random_state=0)
print(df.head)

idg = ImageDataGenerator(rescale=1. / 255.)
train_df, test_df = train_test_split(df, test_size=0.3, shuffle=False)

#print(df["class"].unique())
#print(df["class"].nunique())
#print(train_df.groupby(["class"]).count())
#print(test_df.groupby(["class"]).count())
# print(train_df.head)
# print(test_df.head)

#Batch size and separating training and testing data
batch_size_in = 30
train_ds = idg.flow_from_dataframe(dataframe=train_df, directory=base_path2, batch_size=batch_size_in, shuffle=False)
test_ds = idg.flow_from_dataframe(dataframe=test_df, directory=base_path2, batch_size=batch_size_in, shuffle=False)
```

```python
#######################################MODEL CREATION###################################################

n_epochs = 10
DropoutValue = 0.8
pix_size = 60
learning_rate = 0.00085

# Dropout layers have a seed value that removes the randomness in the outputs
model = tf.keras.models.Sequential()
#The Lambda layer resizes the image to speed up the process
model.add(keras.layers.Lambda(lambda image: tf.image.resize(image, (pix_size, pix_size))))
model.add(keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(pix_size, pix_size, 3)))
model.add(keras.layers.MaxPooling2D((2, 2)))
model.add(tf.keras.layers.Dropout(DropoutValue))
model.add(keras.layers.Conv2D(64, (5, 5), activation='relu'))
model.add(keras.layers.MaxPooling2D((2, 2)))
model.add(tf.keras.layers.Dropout(DropoutValue))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(256, activation='relu'))


# Softmax in the last layer allows for probabilities to be assigned to each of the category entries
model.add(keras.layers.Dense(8, activation='softmax'))

loss_fn = keras.losses.CategoricalCrossentropy()
optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
model.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])
```

```python
######################GENERATE DATA AND RUN NETWORK#############################

# Adding a time of generation for the separate logs
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir="./logs/logs" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
# saving model weights
checkpoint_path = "model_weights/cp-{epoch:04d}.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)
cp_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_path,
    verbose=1,
    save_weights_only=True,
    #saves every epoch
    save_freq="epoch")

#Can save model weights if required using callbacks
#history = model.fit(train_ds, validation_data=test_ds, batch_size=batch_size_in, epochs=n_epochs,
#                    callbacks=[tensorboard_callback, cp_callback,tf.keras.callbacks.EarlyStopping(monitor="val_loss",patience=4)])
```

```python
#######################Train network##############################
history = model.fit(train_ds, validation_data=test_ds, batch_size=batch_size_in, epochs=n_epochs)
model.summary()
```

```python
########################LOAD WEIGHTS AND MAKE PREDICITIONS#####################

# # loading weights
# checkpoint_path = "model_weights/cp-{epoch:04d}.ckpt"
# checkpoint_dir = os.path.dirname(checkpoint_path)
# latest = tf.train.latest_checkpoint(checkpoint_dir)
# model.load_weights(latest)
```

```python
############################ PREDICT ####################################################
predicted = model.predict(test_ds)
#Note: Final batch only contains 8 entries
```

```python
################### Training and test accuracy as a function of training epochs #########
keys = history.history.keys()
print("\033[1mTraining history contains the following keys:\033[0m")
for key in keys:
    print("\t", key)

print("\n\033[1mDisplay the evolution of the accuracy as a function of the training epoch\033[0m\n")
print("  N(Epochs)           = ", n_epochs)
print("  accuracy (train)     = ", history.history['accuracy'])
print("  accuracy (validate)  = ", history.history['val_accuracy'])
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validate'], loc='lower right')
plt.show()
plt.clf()

print("\n\033[1mDisplay the evolution of the loss as a function of the training epoch\033[0m\n")
print("  N(Epochs)        = ", n_epochs)
print("  loss (train)     = ", history.history['loss'])
print("  loss (validate)  = ", history.history['val_loss'])

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validate'], loc='upper right')
plt.show()
```

```python
###### Finding 10 successful predictions and 10 non-succesful predictions #####
#### condition2 checks that we have at least two of each target class in successful cases, from the whole test dataset### 
list_of_ten_successes = []
list_of_ten_fails = []    

for bnum in range(0, len(test_ds)):
    for a in range(0, len(test_ds[bnum][1])):
        target_val = np.argmax(test_ds[bnum][1][a])
        pred_val = np.argmax(predicted[a + bnum * batch_size_in])
        if target_val == pred_val:
            list_of_ten_successes.append([a + bnum * batch_size_in, target_val, pred_val, a, bnum])
        else:
            list_of_ten_fails.append([ a + bnum * batch_size_in, target_val, pred_val, a, bnum])
            
        condition = (len(list_of_ten_fails) >= 10) and (len(list_of_ten_successes) >= 10)
        condition2 = all([[list_of_ten_successes[i][2] for i in range(len(list_of_ten_successes))].count(j)>=1 for j in range(0,8)])
        if condition and condition2:
            break
    if condition and condition2:
        break
```

```python
#List of each target class success
per_class = []
class_List = [0, 1, 2, 3, 4, 5, 6, 7]
for a in class_List:
    per_class.append([x for x in list_of_ten_successes if x[1] == a])
#print(per_class)

#Extracting 10 successes, with at least one from each class
new_list = []
index = 0
while len(new_list) < 10:
    for sublist in per_class:
        if index < len(sublist):
            new_list.append(sublist[index])
        if len(new_list) == 10:
            break
    index += 1

#combining with failures to create a list with 20 data: 10 success (with at least one of each target class) 10 fails
#print(len(new_list))
twenty_list = new_list + list_of_ten_fails[0:10]
random.shuffle(twenty_list)
print(twenty_list)
print(len(twenty_list))
```

```python
####################################  Plot of the 20 images #####################################

# 0,other
# 1,crater
# 2,dark dune
# 3,slope streak
# 4,bright dune
# 5,impact ejecta
# 6,swiss cheese
# 7,spider

class_dict = {0: "Other", 1: "Crater", 2: "Dark Dune", 3: "Slope Streak", 4: "Bright Dune", 5: "Impact Ejecta",
              6: "Swiss Cheese", 7: "Spider"}

plt.figure(figsize=(8, 8))
for i in range(0, len(twenty_list)):
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(test_ds[twenty_list[i][4]][0][twenty_list[i][3]], cmap="Reds")
    plt.xlabel("T: " + str(class_dict[twenty_list[i][1]]) + "\n" + " P: " + str(class_dict[twenty_list[i][2]]))
    plt.tight_layout(pad=1.08, h_pad=None, w_pad=None, rect=None)
plt.show()
```

# Additional code - build confusion matrix #

```python
################# Creating target/true and prediction lists for all data and generating data input for 20 plot, and making a confusion matrix##########

all_pred_list = []
target_list = []
pred_list = []


for bnum in range(0, len(test_ds)):
    print(bnum)
    for a in range(0, len(test_ds[bnum][1])):
        target_list.append(np.argmax(test_ds[bnum][1][a]))
        pred_list.append(np.argmax(predicted[a + bnum * batch_size_in]))
        all_pred_list.append([a + bnum * batch_size_in, np.argmax(test_ds[bnum][1][a]),
                              np.argmax(predicted[a + bnum * batch_size_in]), a, bnum])
```

```python
###########Creating a list of 20 images with at least one from each class and no repeats.###########################

# Creating a list of class lists. x[1] is the true class label which we organise by.
class_List = [0, 1, 2, 3, 4, 5, 6, 7]
per_class = []
for a in class_List:
    per_class.append([x for x in all_pred_list if x[1] == a])


copy_per_class = per_class
twenty_list = []
count = 0
count_val = 20

while count < count_val:
    for cat in range(0, len(per_class)):
        twenty_list.append(copy_per_class[cat][0])
        copy_per_class[cat].remove(copy_per_class[cat][0])
        count += 1
        if count >= count_val:
            break
    if count >= count_val:
        break
print("Twenty image list length: ", len(twenty_list))
print("Image list (twenty_list) data: ", twenty_list)
```

```python
####################################  Plot of 20 images #####################################

# 0,other
# 1,crater
# 2,dark dune
# 3,slope streak
# 4,bright dune
# 5,impact ejecta
# 6,swiss cheese
# 7,spider

class_dict = {0: "Other", 1: "Crater", 2: "Dark Dune", 3: "Slope Streak", 4: "Bright Dune", 5: "Impact Ejecta",
              6: "Swiss Cheese", 7: "Spider"}

plt.figure(figsize=(8, 8))
for i in range(0, len(twenty_list)):
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(test_ds[twenty_list[i][4]][0][twenty_list[i][3]], cmap="Reds")
    plt.xlabel("T: " + str(class_dict[twenty_list[i][1]]) + "\n" + " P: " + str(class_dict[twenty_list[i][2]]))
    plt.tight_layout(pad=1.08, h_pad=None, w_pad=None, rect=None)
plt.show()
```

```python
######################Generate the confusion matrix from targets and preds ##############
cm = confusion_matrix(target_list, pred_list, normalize='true')
display_labels = np.unique(np.concatenate((target_list, pred_list)))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=display_labels)
fig, ax = plt.subplots(figsize=(8,8))
disp.plot(ax=ax,cmap='Blues', xticks_rotation='vertical',colorbar=False)
cax = fig.add_axes([ax.get_position().x1+0.01,ax.get_position().y0,0.02,ax.get_position().height])
plt.colorbar(disp.im_,  cax=cax,cmap='Blues')
plt.show()
```
